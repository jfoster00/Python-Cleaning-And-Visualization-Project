## Dataset Overview

This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area called "Ford GoBike". Ford Gobike was created by the Ford motor company in partnership with the company "Motivate" and introduced to the San Fransisco Bay Area in 2017. The goal of the system is to help reduce traffic congestion in the city by providing an alternative mode of transportation. More information on this system can be located on Ford's media website [HERE](https://media.ford.com/content/fordmedia/fna/us/en/news/2017/06/27/ford-gobike-launching-in-bay-area-bike-sharing.html).

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
trips= pd.read_csv('trips.csv')
```

```python
def uni_pie(column,title):
    # This will speed creating univariate pie charts
    sorted_counts= trips[column].value_counts()
    plt.figure(figsize=(9,7))
    plt.pie(sorted_counts,
            labels= sorted_counts.index,
            autopct='%1.0f%%',
            pctdistance= .8,
            shadow= False,
            startangle = 90,
            counterclock = False,
            wedgeprops={"edgecolor":"black",'linewidth': .5, 'antialiased': True});
    plt.title(title, fontsize= 15);
```

```python
def bi_clusterbar(df,x,hue,title,xlab,ylab):
    # Used to create grouped box plots with two categorical data variables  quicker
    base_color = sb.color_palette()[0]
    plt.figure(figsize=(9,7))
    sb.countplot(data= df, x= x, hue= hue)
    plt.title(title)
    plt.xlabel(xlab)
    plt.ylabel(ylab);
```

## (Types of Users Across All Trips

Below is a breakdown of user types across the full user base. We can see that across all trips available in the data set; 91% of the trips were taken by Subscriber users and 9% of the trips were taken by Customer users.

```python
uni_pie('user_type',
        'Types of Users Across All Trips')
```

## Gender of Users Across All Trips

Below is a breakdown of gender across the full user base. We can see that across all trips available in the data set; 75% of the trips were taken by male users, 23% of trips were taken by female users, and 2% of trips were taken by users of other genders.

```python
uni_pie('gender',
        'Gender of Users Across All Trips')
```

## User Type by Gender Composition
Below we have integrated our user type and gender breakdowns to show the overall gender composition inside both the Subscriber and Customer user types. This gives us a more clear visual representation that validates our findings in the previous slides. Customer users comprise about one tenth of the population of Subscriber users in each gender category and we see that female users combined with other gender users combined are about one third the population of male users.

```python
bi_clusterbar(trips,
              'user_type',
              'gender',
              'User Type By Gender Composition',
              'Type of User',
              'Number of Users')
```

## Average Trip Time for Each User Type by Gender

Below we begin comparing the different user types, seperated by genders, based on average number of minutes they take per trip. It is interesting to note that even though many more trips were taken by Subscriber users, it appears they have a lower average trip time in minutes in every gender category as opposed to the trips taken by Customer users.  

**Subscribers (least to most):**
*   Male: About 9 minutes. 
* Female: About 10 minutes.
*  Other: About 11.5 minutes.

**Customers (least to most):**
*  Other: About 14.7 minutes. 
*   Male: About 15 minutes.
* Female: About 17 minutes.

```python
plt.figure(figsize=(7,9))
sb.catplot(data= trips, x= 'user_type', y= 'minutes', hue= 'gender', kind='bar')
plt.title('Average Trip Time for Each User Type by Gender')
plt.xlabel('Type of User')
plt.ylabel('Average Trip Time in Minutes');
```

## Average Age for Each User Type by Gender

Below we end the presentation by comparing the different user types, seperated by genders, based on their average age in years. It is interesting to note that even though Subscribers and Customers take a largerly different amount of trips, their average ages across genders is nearly identical.

```python
plt.figure(figsize=(7,9))
sb.catplot(data= trips, x= 'user_type', y= 'age', hue= 'gender', kind='bar')
plt.title('Average Age for Each User Type by Gender')
plt.xlabel('Type of User')
plt.ylabel('Average Age in Years');
```
